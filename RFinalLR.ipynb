Here, we start by reading the dataset and setting to a variable to call upon it later

```r
library(tidyverse)
```

```r
df <- read.csv("regrex1.csv", header=TRUE, stringsAsFactors=FALSE)
```

```r
sample <- df %>% select('x','y')
```

Here, we plot the raw data as a scater plot

```r
plt1 <- ggplot(data = df, mapping = aes(x = x, y = y)) + geom_point() + labs(x = "x", x = "y") + theme(axis.title = element_text(size = 20))
```

```r
plt1
```

Here, we create the regression model and display a few key values

```r
regr <- lm(y ~ x, data = df)
```

```r
print(regr)
```

```r
corr_val <- df %>% summarize(correlation = cor(x,y))
```

```r
print(corr_val)
```

Here, we graph the linear regression model using geom_smooth rather than creating a set of prediction values.
We plot the model both alone and with the original, raw data.

```r
plt2 <- ggplot(data = df, mapping = aes(x = x, y = y)) + geom_smooth(method = "lm", se = FALSE) + labs(x = "x", x = "y") + theme(axis.title = element_text(size = 20))
```

```r
plt2
```
